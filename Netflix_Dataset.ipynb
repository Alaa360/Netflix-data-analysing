```python
df = pd.read_csv('Netflix_dataset.csv')
```

### Checking if there is any duplicate, and remove it

```python
df[df.duplicated() == True]
```

```python
df.drop_duplicates()
```

### Checking if there is any null value

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), vmin=0, vmax=1)
```

### The (Show_id) and (Director) for 'House of Cards'

```python
df[df['Title'].isin(['House of Cards'])]
```

##### OR

```python
df[df['Title'].str.contains(['House of Cards'])]
```

### The highest number of movies and TV shows released in a Year

```python
df['Date_N'] = pd.to_datetime(df['Release_Date'])
```

```python
df['Year'] = df['Date_N'].dt.year.astype('Int64')
df['Year'].value_counts()
```

```python
plt.style.use('ggplot')
df['Date_N'].dt.year.astype('Int64').value_counts().plot(kind='bar')
```

### Number of TV shows and Movies are in the Dataset

```python
df.groupby(['Category']).Category.count()
```

```python
sns.countplot(data=df, x = 'Category')
```

### All Movies that released in 2020

```python
df[(df['Year'] == 2020) & (df['Category'] == 'Movie')]
```

### Show all the Titles of all TV shows that are released in India

```python
df[(df['Category'] == 'TV Show') & (df['Country'] == 'India')].Title
```

### Top 10 Directors that have the highest number of movies and tv shows

```python
df['Director'].value_counts().head(10)
```

### Show all the Records that their (Category) is Movie and the (Type) is Comedies or (Country) is UK

```python
df[(df['Category'] == 'Movie') & (df['Type'].str.contains('Comedies')) | (df['Country'].str.contains('United Kingdom'))]
```

### Movies and TV shows that Tom Cruise was in the cast

```python
df[df['Cast'].str.contains('Tom Cruise') == True]
```

### TV Show got the 'R' rating after 2018

### The maximum duration of a Movie/Show on Netflix

```python
df.Duration.unique()
```

```python
df[['Minutes', 'Unit']] = df['Duration'].str.split(' ', expand=True)
df.head(5)
```

```python
df['Minutes'] = pd.to_numeric(df['Minutes'])
```

```python
df.groupby(['Category', 'Unit'])['Minutes'].max()
```
